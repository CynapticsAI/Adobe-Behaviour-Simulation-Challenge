```python
import requests

# Set a timeout value in seconds
timeout_seconds = 30

# Make a request using the requests library with timeout
input_data = {
    "input": "Hello there!!",           # Message
    "system_prompt":system_prompts2,   # System prompt
    "max_tokens": token_number               # Max new tokens
}
response = requests.post("https://ari9dam-orca-2-13b.hf.space/--replicas/55rr7/chat", json=input_data, timeout=timeout_seconds)

# Check for successful response
if response.status_code == 200:
    result = response.json()
    print(result)
else:
    print(f"Error: {response.status_code}, {response.text}")
```
